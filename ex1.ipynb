```python
import sqlite3
import xmltodict
import requests
from datetime import datetime
```

```python
url = "https://knesset.gov.il/Odata/ParliamentInfo.svc/KNS_IsraelLawMinistry()"
xml_url = requests.get(url)
dict = xmltodict.parse(xml_url.content)
db = sqlite3.connect("my_database.db")
cursor = db.cursor()
```

```python
cursor.execute('''
CREATE TABLE KNS_IsraelLawMinistry(
LawMinistryID INTEGER PRIMARY KEY,
IsraelLawID INTEGER,
GovMinistryID INTEGER,
LastUpdatedDate DATETIME2)
''')
db.commit()
```

```python
for line in dict['feed']['entry']:
    data = line['content']['m:properties']
    LawMinistryID = data['d:LawMinistryID']['#text']
    IsraelLawID = data['d:IsraelLawID']['#text']
    GovMinistryID = data['d:GovMinistryID']['#text']
    LastUpdatedDate = data['d:LastUpdatedDate']['#text']

    params = [LawMinistryID, IsraelLawID, GovMinistryID, LastUpdatedDate]
    cursor.executemany('''
    INSERT INTO KNS_IsraelLawMinistry(LawMinistryID, IsraelLawID, GovMinistryID, LastUpdatedDate) VALUES (?,?,?,?)''', [params])
    db.commit()
```

```python
cursor.execute('''SELECT * FROM KNS_IsraelLawMinistry''')
all_rows = cursor.fetchall()
print()
for row in all_rows:
    print('{0} : {1}, {2}, {3}'.format(row[0], row[1], row[2], row[3]))
```
